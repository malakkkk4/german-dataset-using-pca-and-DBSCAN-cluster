```python
selected_cols=["Age","Credit amount","Duration"]
cluster_data=data.loc[:,selected_cols]
```

```python
def distributions(df):
  fig,(ax1,ax2,ax3)=plt.subplots(3,1,figsize=(8,8))
  sns.distplot(df['Age'],ax=ax1)
  sns.distplot(df['Credit amount'],ax=ax2)
  sns.distplot(df['Duration'],ax=ax3)
  plt.tight_layout()
```

```python
distributions(cluster_data)
```

```python
cluster_log=np.log(cluster_data)
distributions(cluster_log)
```

```python
scaler=StandardScaler()
cluster_scaled=scaler.fit_transform(cluster_log)
cluster_scaled
```

```python
wcss=[]
K=range(1,15)
for k in K:
  km=KMeans(n_clusters=k)
  km=km.fit(cluster_scaled)
  wcss.append(km.inertia_)
plt.figure(figsize=(15,5))
plt.plot(K,wcss,'bx-')
plt.xlabel('k');
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
scores = []
for k in range(2,15):
  kmeans=KMeans(n_clusters=k,random_state=0)
  kmeans.fit(cluster_scaled)
  scores.append(silhouette_score(cluster_scaled,kmeans.labels_))
  print("for n_clusters ={}, silhouette score is {}".format(k,silhouette_score(cluster_scaled,kmeans.labels_)))
```

```python
kmeans_sel = KMeans(n_clusters=3, random_state=1).fit(cluster_scaled)
labels = pd.DataFrame(kmeans_sel.labels_)
clustered_data = cluster_data.assign(Cluster=labels)
```

```python
from sklearn.cluster import DBSCAN
outlier_percent= []
for eps in np.linspace(0.1,3,50):
  dbscan=DBSCAN(eps=eps)
  dbscan.fit(cluster_scaled)
  perc_outliers=100* np.sum(dbscan.labels_==-1)/len(dbscan.labels_)
  outlier_percent.append(perc_outliers)
```

```python
sns.lineplot(x=np.linspace(0.1,3,50),y=outlier_percent)
plt.xlabel('Epsilon')
plt.ylabel('Percentage of Outliers')
plt.show()
```
