```python
# importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
df=pd.read_csv('/content/german_credit_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Saving accounts'] = df['Saving accounts'].fillna('Unknown')
df['Checking account'] = df['Checking account'].fillna('Unknown')
```

```python
df.info()
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
```

```python
for n in numerical_cols:
  SC=StandardScaler()
  df[n]=SC.fit_transform(df[[n]])
```

```python
for c in categorical_cols:
  le=LabelEncoder()
  df[c]=le.fit_transform(df[c])
```

```python
df
```

```python
sns.clustermap(df.corr(),annot=True)
plt.show()
```

```python
outlier_percent = []

for eps in np.linspace(0.1,3,50):

    # Create Model
    dbscan = DBSCAN(eps=eps)
    dbscan.fit(df)


    # Log percentage of points that are outliers
    perc_outliers = 100 * np.sum(dbscan.labels_ == -1) / len(dbscan.labels_)

    outlier_percent.append(perc_outliers)
#outlier_percent
```

```python
# Create a line plot of the percentage of outlier points versus the epsilon value choice.
sns.lineplot(x=np.linspace(0.1,3,50),y=outlier_percent)
plt.ylabel("Percentage of Points Classified as Outliers")
plt.xlabel("Epsilon Value")
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=1.6, min_samples=20)  # tune eps!
df["DBSCAN_cluster"] = dbscan.fit_predict(df)
```

```python
df
```

```python
df['DBSCAN_cluster'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Credit amount', y='Duration', hue='DBSCAN_cluster', palette='viridis')
plt.title('DBSCAN Clustering Results')
plt.xlabel('Credit amount (Scaled)')
plt.ylabel('Duration (Scaled)')
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=4)
X_pca = pca.fit_transform(df)

print("Explained variance ratio:", pca.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], cmap="rainbow", s=30)
plt.title("DBSCAN Clusters (PCA 2D)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```
